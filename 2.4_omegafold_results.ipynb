````python
from moleculekit.molecule import Molecule
import numpy as np
import glob
import seaborn as sns
import matplotlib.pyplot as plt
````

#Â Obtain predictions

Run on the alex cluster, first by splitting each folder into 20 folders:
    `for i in `seq 1 20`; do mkdir -p "folder$i"; find . -type f -maxdepth 1 | head -n 572 | xargs -i mv "{}" "folder$i"; done`
    Then, by concatenating files inside each folder to a single one:
        ```for i in {1..20};do cd folder${i}; for file in *.fasta;do (echo ${file}; cat "${file}"; echo) >> finalfile.txt;done; cd ../;done```
        
        
        Then, run this run.sh file:
```bash

#SBATCH --time=24:00:00
#SBATCH --gres=gpu:a40:1
#SBATCH --partition=a40

source /home/woody/b114cb/b114cb10/zymCTRL/hface2/bin/activate

```

# Generated - so far

````python
generated_mols = glob.glob('/agh/projects/noelia/NLP/zymCTRL/sequences/omegafold/generated_2/*.pdb')
````

````python
len(generated_mols)
````

````python
betas_generated={}
for index,mol in enumerate(generated_mols):
    print(index)
    gmol=Molecule(mol)
    betas_generated[mol] = np.mean(gmol.beta)
````

````python
print("total mean is", np.mean(list(betas_generated.values())))
````

````python
len([x for x in list(betas_generated.values()) if x >70 ])/len(list(betas_generated.values()))*100
````

````python
x = [i for i,x in enumerate(betas_generated.values())]
y = [x for x in betas_generated.values()]
````

````python
fig = plt.figure(figsize=(12,12))
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.plot(y,'o', color="#b7dfe4",markeredgecolor="#70C0CB", markersize=2, markeredgewidth=0.5)
plt.savefig('/agh/projects/noelia/NLP/zymCTRL/figs/figs3b.png', dpi=1200)
````
