# Compute statistics on predicted structures

```python
generated_2 = glob.glob('/agh/projects/noelia/NLP/zymCTRL/sequences/esmfold/generated_2/*.pdb')
```

```python
len(generated_2)
```

```python
betas_generated={}
for index,mol in enumerate(generated_2):
    print(mol)
    mol_file = open(mol)
    header = mol_file.readline()
    if header == 'INTERNAL SERVER ERROR':
        print("INTERNAL SERVER ERROR")
        os.remove(seq_file_name)
        continue
    if header == '{"message": "Endpoint request timed out"}':
        print("TIME OUT")
        continue
    if '{"message":"Forbidden"}' in header:
        print("FORBIDDEN")
        continue        
    if 'Sequence is only allowed to have' in header:
        print("FORBIDDEN")
        continue
    gmol=Molecule(mol)
    betas_generated[mol] = np.mean(gmol.beta)
```

```python
len(betas_generated)
```

```python
print("total mean is", np.mean(list(betas_generated.values())))
```

```python
len([x for x in list(betas_generated.values()) if x >0.70 ])/len(list(betas_generated.values()))*100
```
