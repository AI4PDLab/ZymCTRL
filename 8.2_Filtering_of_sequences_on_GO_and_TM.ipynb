```python
#Convert tsv to csv
import re
 
# reading given tsv file
with open("bfactor_scores_carbonic_anhydrases.tsv", 'r') as myfile: 
  with open("bfactor_scores_carbonic_anhydrases.csv", 'w') as csv_file:
    for line in myfile:
       
      # Replace every tab with comma
      fileContent = re.sub("\t", ",", line)
       
      # Writing into csv file
      csv_file.write(fileContent)
 
# output
print("Successfully made csv file")
```
