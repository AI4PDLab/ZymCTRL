# Prediction for Natural sequences

```python
natural_2 = glob.glob('/agh/projects/noelia/NLP/zymCTRL/sequences/natural_2/*_1.fasta')
```

```python
len(natural_2)
```

```python
for seq_file_name in natural_2:
    seq_file = open(seq_file_name)
    label = seq_file_name.split('/')[-1].split('.fasta')[0].split('_')[0]
    if os.path.isfile(f'/agh/projects/noelia/NLP/zymCTRL/sequences/esmfold/natural_2/{label}_1.pdb'):
        print("file exists")
        continue
    sequence = seq_file.readline()
    if len(sequence) > 400:
        print("Sequence is longer than 400")
        continue
    print(f"predicting sequence {label}")
    res = requests.post(url, data=sequence)
    pdb = open(f'/agh/projects/noelia/NLP/zymCTRL/sequences/esmfold/natural_2/{label}_1.pdb','w')
    pdb.writelines(res.text)
    pdb.close()
```

# Stastitics on predicted structures

```python
natural_2 = glob.glob('/agh/projects/noelia/NLP/zymCTRL/sequences/esmfold/natural_2/*.pdb')
```

```python
len(natural_2)
```

```python
betas_natural={}
for index,mol in enumerate(natural_2):
    print(mol)
    mol_file = open(mol)
    header = mol_file.readline()
    if header == 'INTERNAL SERVER ERROR':
        print("INTERNAL SERVER ERROR")
        continue
    if header == '{"message": "Endpoint request timed out"}':
        print("TIME OUT")
        continue
    if '{"message":"Forbidden"}' in header:
        print("FORBIDDEN")
        continue
    if 'Sequence is only allowed to have' in header:
        print("FORBIDDEN")
        continue
    gmol=Molecule(mol)
    betas_natural[mol] = np.mean(gmol.beta)
```

```python
len(betas_natural)
```

```python
print("total mean is", np.mean(list(betas_natural.values())))
```
