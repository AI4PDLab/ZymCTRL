```python
import pandas as pd 
```

```python
#Convert tsv to csv
import re
 
# reading given tsv file
with open("/scratch2/geraldenemunsamy/carbonic_anhydrase/sequence_selection/output_23_01_23.tsv", 'r') as myfile: 
  with open("/scratch2/geraldenemunsamy/carbonic_anhydrase/sequence_selection/output_23_01_23.csv", 'w') as csv_file:
    for line in myfile:
       
      # Replace every tab with comma
      fileContent = re.sub("\t", ",", line)
       
      # Writing into csv file
      csv_file.write(fileContent)
 
# output
print("Successfully made csv file")
```
