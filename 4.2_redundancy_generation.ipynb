This script looks at whether generation for very redundant classes, produces high-identity sequences.

```python
import os
```

#Â First: Compute redundancy in the natural classes

```python
natural_sequences = {}
brenda = open('/agh/projects/noelia/NLP/zymCTRL/dataset_preparation/brenda_uniprot.txt', 'r')
for line in brenda:
    columns = line.split('\t')
    key = columns[5]
    if key not in natural_sequences:
        natural_sequences[key] = []
        natural_sequences[key].append(columns[6].strip())
    else:
        natural_sequences[key].append(columns[6].strip())        
brenda.close()
```

```python
natural_sequences.keys()
```

```python
for key, sequences in natural_sequences.items():
    if key == 'EC number': 
        continue
    if ';' in key: 
        continue
    if key == '':
        continue
    print(key, len(sequences))
    if len(sequences) > 300:
        continue
    path = f'/agh/projects/noelia/NLP/zymCTRL/redundancy/{key}/'
    if os.path.exists(path):
        continue
    else:
        os.mkdir(path)
    for index,sequence in enumerate(sequences):
        fn = open(f'{path}/{key}_{index}.fasta','w')
        fn.write(f'>{key}_{index}\n')
        fn.write(f'{sequence}\n')
        fn.close()
        
```
