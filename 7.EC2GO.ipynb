```python
import pandas as pd 
```

```python
#Read in EC2GO
df = pd.read_csv('ec2go.tsv', sep='\t', header=None)
```

```python
#Split dataframe on ; 
df_2 = df[0].str.split(';', expand=True)
```

```python
#Delete columns that will not be needed
df_2 = df_2.drop([1, 3], axis=1)
```

```python
df_2
```

```python
n=2
df_3 = df_2.tail(df.shape[0] -n)
df_4 = df_3.rename(columns = {0: 'EC_number', 2: 'GO'})
df_4['GO'] = df_4['GO'].str.replace(" ", "")
```

```python
#EC numbers and respective GO term 
df_4
```

```python
df_4.to_csv('ec_2_go_cleaned.csv', index=False)
```

```python
#Convert tsv to csv
import re
 
# reading given tsv file
with open("zymcontrolclean.faa_ips_results_final.tsv", 'r') as myfile: 
  with open("zymcontrolclean.faa_ips_results_final.csv", 'w') as csv_file:
    for line in myfile:
       
      # Replace every tab with comma
      fileContent = re.sub("\t", ",", line)
       
      # Writing into csv file
      csv_file.write(fileContent)
 
# output
print("Successfully made csv file")
```
