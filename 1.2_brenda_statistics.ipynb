```python
len_dis_processed = {
    'EC number': [x for x in len_dis.keys() if ';' not in x],
    'Frequencies': [value for x,value in len_dis.items() if ';' not in x],
    'EC class': [x.split('.')[0] for x in len_dis.keys() if ';' not in x]
}
df = pd.DataFrame(len_dis_processed)
```

```python
df
```

```python
print("Total number of labels is", len(len_dis_processed['EC number']))
print("Total number of labels with less or 5 sequences", 
      len([x for x in len_dis_processed['Frequencies'] if x<=5]))
print("Total number of labels with less or 1 sequences", 
      len([x for x in len_dis_processed['Frequencies'] if x<=1]))
print("Percentage of labels with less or 5 sequences", 
      len([x for x in len_dis_processed['Frequencies'] if x<=5])/len(len_dis_processed['Frequencies'])*100)
print("Percentage of labels with less or 1 sequences", 
      len([x for x in len_dis_processed['Frequencies'] if x<=1])/len(len_dis_processed['Frequencies'])*100)
```

```python
a=[x for x in len_dis_processed['Frequencies']]
a.sort()
print("Coverage of the 100 largest classes is", sum(a[-100:])/sum([x for x in len_dis_processed['Frequencies']])*100)
```
