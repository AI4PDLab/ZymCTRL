```python
import pandas as pd
from argparse import ArgumentParser
from tqdm import tqdm
import pickle
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer, AutoTokenizer
import random
from selenium import webdriver
from selenium.webdriver.common.by import By
import os
import time
from tqdm import tqdm
from selenium.webdriver.firefox.options import Options
import math
#from webdriver_manager.firefox import GeckoDriverManager
```

```python
df = pd.read_csv('/agh/projects/noelia/NLP/zymCTRL/dataset_preparation/brenda_uniprot.txt',sep='\t')
```

```python
EC_cat = ['1','2','3','4','5','6','7']
sub_dict = {}
for i in EC_cat:
    sub = df[df.iloc[:,-2].apply(lambda x: str(x).split('.')[0]) == i]
    lab = sub.iloc[:,-2].unique()
    for j in tqdm(lab):
        data =  sub[sub.iloc[:,-2]==j].sample(n=1,random_state=73)
        sub_dict[j] =  data.iloc[:,-1]
```

```python
with open('natural_every_cat.pickle', 'wb') as handle:
    pickle.dump(sub_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
